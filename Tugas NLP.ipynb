```python
import os
import nltk
import nltk.corpus
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords
```

```python
file = open("debat1.txt", 'r')
debat = file.read()
debat
```

```python
#Tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
debat_tokens = word_tokenize(debat)
debat_slice = debat_tokens[:400]
#debat_tokens
freqdist = FreqDist()
for i in debat_slice:
    freqdist[i.lower()]+=1
freqdist

stop_words = set (stopwords.words("english"))
filtered = [i for i in freqdist if not i in stop_words]
filtered
```

```python

#Stemming
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import re
word_lem = WordNetLemmatizer()
filtered1=[]
for i in filtered:
        filtered1.append(word_lem.lemmatize(i, "v"))
        
punct = re.compile(r'[-.?!,:;()|0-9]')
post_punctuation = []
for i in filtered1:
    word = punct.sub("", i)
    if len(word) > 0:
        post_punctuation.append(word)

freqdist = FreqDist()
for i in post_punctuation:
    freqdist[i.lower()]+=1
print("Kata yang sering muncul: \n {}".format(freqdist.most_common(5)))
```

```python
def get_sumary(debat, pct):
    summary = summarize(debat, ratio=pct, split=True)
    return summary
print("Summary: \n {}".format(get_sumary(debat, 0.3)))
```
